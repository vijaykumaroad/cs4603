```python
%run ../shared_notebook_setup.py
import mlflow

# Specify the tracking server URI, e.g. http://localhost:5000
mlflow.set_tracking_uri("http://127.0.0.1:5000")
mlflow.set_experiment("mlflow-openai-api-tracing")

print("MLflow experiment set: mlflow-openai-api-tracing")
```

## MLFlow Terminilogy

* Trace = full request timeline (your program start --> to finish)
* Spans = timeline segments for each operation (i.e., function)

```python
# The decorator below (`@mlflow.trace`) tells MLflow to automatically trace this function call whenever it runs.
# MLflow span captures execution metadata (for example timing, inputs, outputs, and errors)
# and links it to the active run/trace context.
# span_type is used for grouping the tracing results in the UI.

@mlflow.trace(name="square_plus_one_span", span_type="CHAIN")
def square_plus_one(x: int):
    return x * x + 1

with mlflow.start_run(run_name="mlflow_tracing") as run:
    result = square_plus_one(4)
    mlflow.log_metric("result", result)
    print("Result:", result)
    print("Run ID:", run.info.run_id)
    print("Trace ID:", mlflow.get_last_active_trace_id())
```
